# Equations of one dimentional motion

Instanteanous speed is defined by the specific displacement in a speccific time time ratio like as:
 
\begin{equation}
v = \dfrac{\Delta x}{\Delta t} = \dfrac{x_2-x_1}{t_2-t_1} 
\end{equation}

so, the change of velocity as a function of time define acceleration (instantanoeus) , then :
\begin{equation}
a = \dfrac{\Delta v}{\Delta t} = \dfrac{v_2-v_1}{t_2-t_1} 
\end{equation}

## Pyhton code 
We first must be to call  necessary packages

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import matplotlib as mpl
from IPython.display import display, Math, Latex

# Plots parameters 
mpl.rcParams['figure.dpi']= 180
plt.style.use('science')
plt.rcParams.update({
       "font.size":10})  
```

### Plot Distance vs Time 

```python
# We create timne and  position lists
time = np.arange(0,65,5)
position = np.array([0,4.1,7.9,12.1,16.0,16.0,16.0,18.0,20.1,21.9,24.0,22.1,20.0])

plt.figure(figsize=(5,3))
plt.plot(time,position,'o-b',mfc='white',ms=5)
plt.xlabel('Time (s)')
plt.ylabel('Distance (cm)')
plt.title('Distance vs Time')
plt.show()
```

### Compute instantanous speed 
From eq.(1) we can compute the instantanoeus speed in each instant of time, then :
\begin{equation}
v_{1\to 2} = \dfrac{x_2-x_1}{t_2-t_1} = \dfrac{4.1-0}{5-0} = 0.82 \text{cm/s}
\end{equation}


```python
x2=4.1;x1=0
t2=5;t1=0

v12 = (x2-x1)/(t2-t1)
print('v = %f'%v12)
```

```python
v =np.zeros((len(time),2))
for i in range(1,len(time)):
    deltat = time[i]-time[i-1]
    deltax = position[i]-position[i-1]
    velocity = deltax/deltat
    v[i-1,1]=velocity
    v[i,0]=time[i]
plt.figure(figsize=(5,3))
plt.plot(v[:,0],v[:,1],'o-r',mfc='white',ms=5)
plt.plot([0,60],[0,0],'--k',lw=1)
plt.xlabel('Time (s)')
plt.ylabel('Speed (cm/s)')
plt.xlim([-.5,60.5])

plt.show()
```

### Compute instantanous acceleration
From eq.(2) we can compute the instantaneous acceleration in each instant of time, then :
\begin{equation}
a_{1\to 2} = \dfrac{v_2-v_1}{t_2-t_1} = \dfrac{5-0}{4.1-0} = -0.012000 \approx 0 \text{cm/s}
\end{equation}


```python
v2=v[1,1];v1=v[0,1]
t2=time[1];t1=time[0]

a12 = (v2-v1)/(t2-t1)
print('a = %f'%a12)
```

```python
a =np.zeros((len(time),2)) # aceleration 1D array
for i in range(1,len(time)):
    deltat = time[i]-time[i-1]
    deltav = v[i,1]-v[i-1,1]
    aceleration = deltav/deltat
    a[i-1,1]=aceleration
    a[i,0]=time[i]
    
# graphics    
plt.figure(figsize=(5,3))
plt.plot(a[:,0],a[:,1],'o-g',mfc='white',ms=5)
plt.plot([0,60],[0,0],'--k',lw=1)
plt.xlabel('Time (s)')
plt.ylabel('Aceleration (cm/s$^2$)')
plt.xlim([-0.5,60.5])
plt.show()
```

# Comparative of three plots
